```python
#Ejercicio 1: mostrar las primeras filas del dataframe
print("Ejercicio 1: Mostrar las primeras filas del DataFrame")
print(df.head (5))
```

```python
#Ejercicio 2: Calcular el total de ventas por región
print("Ejercicio 2: Calcular el total de ventas por región")
ventas_por_región = df.groupby('Region')['Ventas'].sum()
print(ventas_por_región)
print("/n")
print(df.head(10))
```

```python
#Ejercicio 3: Filtrar el DataFrame para mostrar solo las ventas realizadas por David
print("Filtrar el DataFrame para mostrar solo las ventas realizadas por David")
ventas_David = df[df['Vendedor'] == 'David']
print(ventas_David)
```

```python
#Ejercicio4:Calcular el promedio de ventas por región
print("Ejercicio4:Calcular el promedio de ventas por región")
ventas_promedio_por_región=df.groupby(df['Region'])['Ventas'].mean()
print(ventas_promedio_por_región)
```

```python
#Ejercicio5:Calcular el promedio de ventas por región con dos decimales
print("Ejercicio4:Calcular el promedio de ventas por región con dos decimales")
ventas_promedio_por_región_decimal=df.groupby(df['Region'])['Ventas'].mean().round(2)
print(ventas_promedio_por_región_decimal)
```

```python
#Ejercicio6:Ordenar el dataframe por ventas en orden descendente
print("Ejercicio6:Ordenar el dataframe por ventas en orden descendente")
df_ordenado = df.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
```

```python
#Ejercicio7:Filtrar las filas con ventas mayores a 10000 y ordenar de mayor a menor
print("Ejercicio7:Filtrar las filas con ventas mayores a 1000 y ordenar de mayor a menor")
df_filtrado_ventas_mayores_10000 = df[df['Ventas'] > 10000]
df_filtrado_ventas_mayores_10000_ordenado = df_filtrado_ventas_mayores_10000.sort_values(by='Ventas', ascending=False)
print(df_filtrado_ventas_mayores_10000)
print("/n")
print(df_filtrado_ventas_mayores_10000_ordenado)
```

```python
#Ejercicio8:Ordenar el dataframe por ventas en orden descendente
print("Ejercicio8:Ordenar el dataframe por ventas en orden descendente")
df_ordenado = df.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
```

```python
#Ejercicio9:Eliminar la columna vendedor
print("Ejercicio9:Eliminar la columna vendedor")
df_sin_vendedor = df.drop(columns=['Vendedor'])
print(df_sin_vendedor)
```

```python
#Ejercicio10:Calcular el total de ventas por región
print("Ejercicio10:Calcular el total de ventas por región")
ventas_total_por_región=df.groupby(df['Region'])['Ventas'].sum()
print(ventas_total_por_región)
```

```python
#Ejercicio11:Calcular el total de ventas por producto
print("Ejercicio11:Calcular el total de ventas por región")
ventas_total_por_producto=df.groupby(df['Producto'])['Ventas'].sum()
print(ventas_total_por_producto)
```
