```python
query = """SELECT stage as number_of_stage, home_team_goal as avg_home_team_goals, away_team_goal as avg_away_team_goals, SUM (home_team_goal + away_team_goal) as total_goals, COUNT(id)
        FROM Match
        GROUP BY season
        ORDER BY season DESC;"""

#No modificar lo siguiente:
stages_by_season = pd.read_sql(query, conn)
stages_by_season
```

```python
query = """SELECT L.id AS Id_League, L.name AS Nombre_League, C.name AS Pais
        FROM League AS L
        JOIN Country AS C ON C.id = L.country_id;"""

country_leagues = pd.read_sql(query, conn)
country_leagues
```

```python
query = """SELECT Match.season, Country.name AS country_name, sum(match.id) AS count_matches
        FROM Match
        JOIN Country ON Country.id = Match.country_id
        WHERE Season = '2008/2009' AND Country.name IN ('Spain', 'Belgium')
        GROUP BY match.id
        ORDER BY count_matches DESC;"""

country_leagues = pd.read_sql(query, conn)
country_leagues
```
